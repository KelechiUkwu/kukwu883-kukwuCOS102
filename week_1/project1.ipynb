```python
### def calculate_finances():
    print("Financial Calculator")
    print("a. Simple Interest")
    print("b. Compound Interest")
    print("c. Annuity Plan")
    choice = input("Select an option (a, b, or c): ").lower()

    if choice == 'a':
        P = float(input("Enter the Principal: "))
        R = float(input("Enter the Rate: ")) / 100
        T = float(input("Enter the Time: "))
        A = P * ( 1 + R * T )
        print("Total Amount: ",A)

    elif choice == 'b':
        P = float(input("Enter the Principal: "))
        R = float(input("Enter the Rate: ")) / 100
        n = int(input("Enter the number of times interest was gotten: "))
        T = float(input("Enter the time: "))
        A = P * (1 + R/n)** (n * T)
        print("Total Amount: ",A)

    elif choice == 'c':
        PMT = float(input("Enter Periodic Payment (PMT): "))
        R = float(input("Enter Annual Interest Rate in %: ")) / 100
        n = int(input("Enter times compounded per year: "))
        t = float(input("Enter time in years: "))

        # Formula: A = PMT * ((1 + r/n)**(n*t) -1) / (r/n))
        rate_per_period = R / n
        A = PMT *(((1 + rate_per_period)**(n * t)-1) / rate_per_period)
        print("Future Value of Annuity: ",A)

    else:
        print("Invalid selection. Please run the program again.")
if __name__ == "__main__":
            calculate_finances()
                  
```
